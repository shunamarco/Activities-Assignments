```python
while True:
    password = input("Enter password: ")

    naay_letter = any(char.isalpha() for char in password)
    naay_number = any(char.isdigit() for char in password)

    if naay_letter and naay_number:
        print("Password accepted.")
        break

    else:
        print("Invalid password. Try again.")

```
