```python
users=[]

while True:
  print("User Management System")
  print("1. Show Users")
  print("2. Add User")
  print("3. Update User")
  print("4. Delete User")
  print("5. Exit")

  sige = input("Pili chuy: ")

  #Show_asdasd
  if sige == "1":
    print("Show Users")
    if not users:
      print("No users to display.")
    else:
      for user in users:
        print(user)

  #Add_haha
  elif sige == "2":
    name=input("Enter name: ")
    users.append(name)
    print("Current users: ")
    for user in users:
      print(user)

  #Update_hehe
  elif sige == "3":
    karaan_name = input("Enter name to update: ")
    if karaan_name in users:
      index = users.index(karaan_name)
      bagong_name = input("Enter new name: ")
      users[index] = bagong_name
      print("User updated.")
      print("Current users: ")
      for user in users:
        print(user)
    else:
      print("User not found")

  #Delete_part
  elif sige == "4":
    impas = input("Enter name to delete: ")
    if impas in users:
      users.remove(impas)
      print("User deleted.")
      print("Current users:")
      for user in users:
        print(user)
    else:
      print("User not found.")

  #Exit
  elif sige == "5":
    print("Exiting.")
    break

  else:
    print("Invalid choice. Please enter a number between 1 and 5.")

```
